```python
import numpy as np
import statsmodels.api as sm
```

```python
N = 1000
X = np.stack([np.random.normal(loc=0, scale=1, size=N) for _ in range(5)], axis=1)
beta = np.arange(1, 6).reshape(5, -1)
Y = X @ beta + np.random.normal(loc=0, scale=1)

model = sm.OLS(Y, X).fit()
model.summary()
```

```python
N = 1000
X = np.stack([np.random.normal(loc=0, scale=1, size=N) for _ in range(5)], axis=1)
beta = np.arange(1, 6).reshape(5, -1)
Y = X @ beta + np.random.normal(loc=0, scale=1)

model = sm.OLS(X[:, 3], X[:, [0, 1, 2, 4]]).fit()
model = sm.OLS(Y, X[:, 3] - model.predict()).fit()
model.summary()
```
