```python
import os
import numpy as np
from PIL import Image
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
n_clusters = 2
```

```python
path = os.path.join("/Users/lpossner/Pictures/Abschlussfeier", "0_Professoren_Rektor_Dekan2022.jpg")
image = Image.open(path)
arr = np.array(image)

image
```

```python
k_means = KMeans(n_clusters=n_clusters)
k_means.fit(arr.reshape(arr.shape[0] * arr.shape[1], -1))
centers = k_means.cluster_centers_
labels = k_means.labels_

arr_clustered = centers[labels].reshape(arr.shape[0], arr.shape[1], arr.shape[2])

image_clustered = Image.fromarray(arr_clustered.astype('uint8'))

image_clustered
```
