```python
import numpy as np
import seaborn as sns
```

```python
from scipy.stats import pearsonr

N = 1000
X = np.random.randn(N)
Y = 0.1 * X + np.random.randn(N)

r, p = pearsonr(X, Y)

print(r)
print(p)

sns.scatterplot(x=X, y=Y)
```

```python
from scipy.stats import spearmanr

N = 1000
X = np.random.randn(N)
Y = 1.0 * np.tanh(X) + 1.0 * 0.0 * np.random.randn(N)

r, p = spearmanr(X, Y)

print(r)
print(p)

sns.scatterplot(x=X, y=Y)
```
